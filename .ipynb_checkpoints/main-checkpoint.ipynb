```python
# import library
import pandas as pd
import numpy as np
```

# Create Messy Data

```python
# Data
data={
    'Name':['Aman','Riya','Karan','Riya',None],
    'Age':[20,21,None,21,22],
    'City':['Delhi','Mumbai','Delhi',None,'Chennai'],
    'Salary':[50000,None,45000,50000,None]
}
```

```python
# create dataframe
df=pd.DataFrame(data)

df
```

# Understanding the Problem

```python
df.info()
```

```python
# checking null values
df.isnull().sum() # how many null values are there in each column
```

```python
# check for duplicates
df.duplicated()
```

```python
# check the datatypes
df.dtypes
```

# Handling The Issues in the Dataset

## 1. Fill/Remove the missing values

```python
# Fill the Numerical Column with its mean value
# Fill the Text column with any placeholder
# Fill the categorical column with its mode value


df['Name']=df['Name'].fillna('Unknown')
df['Age']=df['Age'].fillna(df['Age'].mean())
df['Salary']=df['Salary'].fillna(df['Salary'].mean())
df['City']=df['City'].fillna('Unknown')
```

```python
# # Another Method of Filling the values.
# df.fillna({
#     'Name':'Unknown',
#     'Age':df['Age'].mean(),
#     'Salary':df['Salary'].mean(),
#     'City':'Unknown'
# },inplace=True
#     )
```

```python
#show dataframe
df
```

## 2. Fix the datatype of Age column

```python
df['Age']=df['Age'].astype(int)
```

```python
df.dtypes
```

## 3. Drop Duplicates

```python
df.drop_duplicates(inplace=True)
```

```python
df
```

## 4.Cleaning the Text Data

```python
df['City']=df['City'].str.strip().str.title()
df['Name']=df['Name'].str.strip().str.title()
```
