```python
# importing library
import pandas as pd
```

```python
# Creating Dataframe

data = {
    "Name": ["Aman", "Riya", "Karan", "riya ", "Rahul", "Aman", None],
    "Age": [25, 30, None, 30, 22, 25, 28],
    "City": ["Delhi", "Mumbai", "delhi ", "Mumbai", None, "DELHI", "Chennai"],
    "Department": ["HR", "IT", "Finance", "IT", "HR", "HR", None],
    "Salary": [50000, 60000, -45000, 60000, None, 50000, 70000],
    "Joining_Date": [
        "2022-01-15", "2021-07-10", "2023-03-01",
        "2021-07-10", "2022-11-20", "2022-01-15", "not_available"
    ]
}

df=pd.DataFrame(data)
df
```

# Inspect The Data

```python
df.info()
```

```python
# no of null values in each column
df.isnull().sum()
```

# 1. Cleaning the Text Data

```python
df['Name']=df['Name'].str.strip().str.title()
df['City']=df['City'].str.strip().str.title()
df['Department']=df['Department'].str.strip().str.upper()
```

```python
df
```

# 2. Doing Filtering of invalid data

```python
df=df[df['Age']>0]
df=df[df['Salary']>0]
```

```python
df
```

# 3. Fixing DataTypes

```python
df['Age']=df['Age'].astype(int)
df['Joining_Date']=pd.to_datetime(df['Joining_Date'],errors="coerce")
```

```python
df.dtypes
```

```python
df
```

# 4. Removing duplicates

```python
df.duplicated()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df
```

# 5. Handling missing values

```python
df.dropna(subset='Joining_Date') # joining date was not able to handle 
```
